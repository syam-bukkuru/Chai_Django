# Building CNN - Convolutional Neural Network

##Project Fashion-Classification - End to End
Train Convolution Neural Network on 60,000 Fashion-MNIST Images (data in NP array)

Test Convolution Neural Network on 10,000 Fashion-MNIST Images (data in NP array)

##Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import keras
```

##Load Data

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
X_train[0]
```

```python
class_labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
'''
0 =`T-shirt/top`
1 = Trouser
2 = Pullover
3 = Dress
4 = Coat
5 = Sandal
6 = Shirt
7 = Sneaker
8 = Bag
9 = Ankle boot
'''
```

# Change Dimensions

```python
X_train.ndim
```

```python
X_train.shape
```

```python
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
```

```python
X_train.ndim
```

```python
X_train.shape
```

# Feature Scaling

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

# Split Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=2020)
```

```python
X_train.shape, y_train.shape
```

```python
X_val.shape, y_val.shape
```

# Convolutional Neural Network - Model Building

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(units = 128, activation = 'relu'),
    keras.layers.Dense(units = 10, activation = 'softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=512, verbose=1, validation_data=(X_val, y_val))
```

# Test the **Model**

```python
model.predict(np.expand_dims(X_test[0], axis=0)).round(2)
```

```python
np.argmax(model.predict(np.expand_dims(X_test[0], axis=0)).round(2))
```

```python
y_test[0]
```

```python
y_pred = model.predict(X_test).round(2)
y_pred
```

```python
model.evaluate(X_test, y_test)
```

# Visualise Output

```python
plt.figure(figsize=(16, 16))

j = 1
for i in np.random.randint(0, 1000, 25):
    plt.subplot(5, 5, j)
    j += 1
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
    plt.title(f"Actual: {class_labels[y_test[i]]}\nPredicted: {class_labels[np.argmax(y_pred[i])]}")

plt.tight_layout()
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16, 9))
y_pred_labels = [ np.argmax(label) for label in y_pred]
cm = confusion_matrix(y_test, y_pred_labels)

#show cm
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_labels, target_names=class_labels))
```

# Convolutional Neural Network - Building Complex Model

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPool2D, Flatten, Dropout, BatchNormalization

cnn_model2 = Sequential([
    Input(shape=(28,28,1)),
    Conv2D(32, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPool2D((2,2)),

    Conv2D(64, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPool2D((2,2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])
```

```python
#compile the model
cnn_model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
cnn_model2.fit(X_train, y_train, epochs=10, batch_size=512, verbose=1, validation_data=(X_val, y_val))
```

# Testing the Model

```python
y_pred = cnn_model2.predict(X_test).round(2)
y_pred
```

```python
cnn_model2.evaluate(X_test, y_test)
```

# Visualise Output

```python
plt.figure(figsize=(16, 16))

j = 1
for i in np.random.randint(0, 1000, 25):
    plt.subplot(5, 5, j)
    j += 1
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
    plt.title(f"Actual: {class_labels[y_test[i]]}\nPredicted: {class_labels[np.argmax(y_pred[i])]}")

plt.tight_layout()
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16, 9))
y_pred_labels = [ np.argmax(label) for label in y_pred]
cm = confusion_matrix(y_test, y_pred_labels)

#show cm
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_labels, target_names=class_labels))
```
